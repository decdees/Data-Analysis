```python
import pandas as pd
import datetime
```

```python
df1 = pd.read_csv('Spintly-1.csv') #reading 1st csv file
df2 = pd.read_csv('Spintly-2.csv') #reading 2nd csv file
df = pd.concat([df1, df2], ignore_index=True) #merging both dataframes(files) into one and ignoring indexing
df = pd.DataFrame(df)
```

```python
print(df.info()) 
```

```python
df['PunchTime'] = pd.to_datetime(df['PunchTime']) #coverting the date to datetime
```

```python
df['Month'] = df['PunchTime'].dt.month_name(locale = 'English')#Adding a month column
```

```python
print(df.info())
```

```python
new_df = df.dropna() #Dropping empty cells
```

# 1. Total number of monthly unique members for all locations. Perform this activity for both the access point direction. [Hint: Entry, Exit and Punch time].

## Total monthly unique members for all location with Access points and Punch Time:

```python
ndf = new_df.loc[:, ['Member_name', 'Location', 'accessPointDirection', 'PunchTime', 'Month']] #Getting required rows into a variable
um = ndf.drop_duplicates(subset = ["Member_name"])#getting distinct Members
um #unique members
```

## Total number of monthly unique members for all locations:

```python
um.groupby(['Month']).size().sort_values(ascending=True)
```

## Unique Members by Entry:


```python
umEntry = um.loc[um.accessPointDirection=='entry'] #unique members by "entry"
umEntry
```

## Total number of monthly unique members for all locations for Entry:

```python
umEntry.groupby(['Month']).size().sort_values(ascending=True)
```

## Unique members by Exit:


```python
umExit = um.loc[um.accessPointDirection=='exit'] #unique members by "exit"
umExit
```

## Total number of monthly unique members for all locations for Exit:

```python
umExit.groupby(['Month']).size().sort_values(ascending=True)
```

# 2.	Total number of monthly unique members.

## Total monthly unique members:

```python
tm = df.loc[:, ['Member_name', 'Month']].dropna()
tmu = tm.drop_duplicates(subset = ["Member_name"])#getting distinct Members
tmu.head(25)
```

# 3.	Total number of monthly unique clients.

## Monthly unique clients:

```python
mclient = df.loc[:, ['Client', 'Month']].dropna() #Getting required rows into a variable
mclientu = mclient.drop_duplicates(subset = ["Client"])#getting distinct Clients
mclientu #unique members
```

## Total number of monthly unique clients:

```python
mclientu.groupby(['Month']).size().sort_values(ascending=True)
```

# 4.	Total number of monthly unique access point names.
