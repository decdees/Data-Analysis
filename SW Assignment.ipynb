```python
import pandas as pd
import datetime
```

```python
df1 = pd.read_csv('Spintly-1.csv') #reading 1st csv file
df2 = pd.read_csv('Spintly-2.csv') #reading 2nd csv file
df = pd.concat([df1, df2], ignore_index=True) #merging both dataframes(files) into one and ignoring indexing
df = pd.DataFrame(df)
```

```python
df['PunchTime'] = pd.to_datetime(df['PunchTime']) #coverting the date to datetime
```

```python
df['Month'] = df['PunchTime'].dt.month_name(locale = 'English')#Adding a month column
```

```python
new_df = df.dropna() #Dropping empty cells
```

# 1. Total number of monthly unique members for all locations. Perform this activity for both the access point direction.

```python
ndf = new_df.loc[:, ['Member_name', 'Location', 'accessPointDirection', 'PunchTime', 'Month']] #Getting required rows into a variable
um = ndf.drop_duplicates(subset = ["Member_name"])#getting distinct Members
um #unique members
```

### Total number of monthly unique members for all locations with PunchTime

```python
monthlyUmembers = df.groupby(['Month', 'Member_name', 'Location', 'accessPointDirection', 'PunchTime']).Month.count()

##exporting to excel file
um = pd.DataFrame(monthlyUmembers)
um.to_csv('MonthlyUniqueMembersAllLocation.csv')

monthlyUmembers
```

### Total number of monthly unique members for all locations without PunchTime

```python
monthlyUmembers = df.groupby(['Month', 'Member_name', 'Location', 'accessPointDirection']).Month.count()

##exporting to excel file
um = pd.DataFrame(monthlyUmembers)
um.to_csv('MonthlyUniqueMembersAllLocations.csv')
um
```

# 2.	Total number of monthly unique members.

```python
tnmu = df.groupby(['Month', 'Member_name']).Month.count()

##exporting to excel file
tnmu = pd.DataFrame(tnmu)
tnmu.to_csv('MonthlyUniqueMembers.csv')

tnmu
```

# 3. Total number of monthly unique clients.

```python
tmuc = df.groupby(['Month', 'Client']).Month.count()

##exporting to excel file
tmuc = pd.DataFrame(tmuc)
tmuc.to_csv('MonthlyUniqueClients.csv')

tmuc
```

# 4.	Total number of monthly unique access point names.

```python
mua = df.groupby(['Month', 'Access_Point_Name']).Month.count()

##exporting to excel file
mua = pd.DataFrame(mua)
mua.to_csv('MonthlyUniqueAccessPointNames.csv')

mua
```
